```python
#Félix Clément

#Cette cellule est le projet final. 
#En dessous de cet cellule, j'ai mis les comparaisons des algorithmes 
#que j'ai modifié par rapport au TP2 (Cela pourrait peut-être m'être utile lors de la présentation du projet)

#Note : plus il y a de cases de jeu, moins les X et les O sont centrés. 
#Je crois que c'est dû que draw de pygame n'accepte pas les floats (Voir les dessins des X et des O dans le Code)

#Bonne Correction ! =) 


import random
import pygame
import math
import time



pygame.init()

#On stock les codes de couleurs dans des variables
noir = (0, 0 ,0)
blanc = (255, 255, 255)
vert =  (0, 255, 0)
rouge = (255, 0, 0)

Taille =[600, 600] #variable qui contient la Taille de la fenêtre de jeu
fenetre = pygame.display.set_mode(Taille)


#Entrée : le nombre de case que contient une ligne
#Rôle : Créer un tableau des positions de jeu de l'ordinateur
#Sortie : Le tableau des positions de jeu de l'ordinateur
def choixCaseOrdi(nbCaseLigne) :
    
     #On crée une liste (initialiser avec des valeurs = 0 pour pouvoir donner une taille) nbCaseLigne est exposé en 2, car ça donne le nombre total de cases
    choixOrdi = [0] * (nbCaseLigne**2)


    j = 0 # Variable pour variation de positionnement X
    k = 0 # Variable pour variation de positionnement Y
    
    #z sert à réduire la distance entre les cases
    z = CalculProportionCase(nbCaseLigne) 
    
    
    i = 0 #Compteur
    
    while i < nbCaseLigne**2 :
            
        #Si le nombre de colonne est dépassé, on remet à 0 pour commencer une nouvelle colonne et on change la rangée
        if j >= nbCaseLigne : 
            j = 0 #On réinitialise j
            k += 1 #On incrémente k pour descendre d'une rangée dans le tableau
                                
        #On dessine la case dans le tableau rect        
        choixOrdi[i] = [(45 + (175 / z) * j), (25 + (175 / z) * k)]
        j += 1  #On augmente le facteur de position de j pour changer de cases
        i += 1
    
    return choixOrdi


#Entrée : 2 list (les choix possibles des cases de jeu)
#Rôle : Permet à l'ordinateur de choisir la case qu'il veut jouer (hasard)
#Sortie : 2 int (son choix de case (Coordonnée X et Y))
def IA_Play(choixOrdi):
    
    #On choisi une valeur aléatoire entre 0 et le total de possibilités -1 (Car on part à 0 au lieu de 1)
    choix = random.randint(0, len(choixOrdi) - 1) #Parfois l'ordinateur peut prendre un peu de temps à jouer, car il tombe sur des cases indisponibles
    
    #Les coordonnées à la position du chiffre aléatoire
    return choixOrdi[choix]


#Entrée : un boolean
#Rôle : Inverse la valeur du boolean (True devient False et False devient True)
#Sortie : Le boolean
def switcher(tourOrdinateur) :
    #Si le bool est faux
    if (tourOrdinateur == False) : 
        tourOrdinateur = True
        return tourOrdinateur
        
    #Si le bool est vrai    
    elif (tourOrdinateur == True) : 
        tourOrdinateur = False
        return tourOrdinateur

    
    
#Entrée : Le nombre de case par ligne   
#Rôle : Fonction qui calcul le z de la fonction tableDeJeu()    
#Sortie : le chiffre calculé par la fonction    
def CalculProportionCase(nbCaseLigne) :
    
    #Table des valeurs possible (Je les ai calculé en faisant des tests)
    # x | 3  |  4   |  5  | 6 |  7   |  8  | 9 |  10  |...
    # y | 1 | 1.25 | 1.6 | 2 | 2.25 | 2.6 | 3 | 3.25 | ...
    
    #À l'aide de cette table de valeur, je peux créer la fonction suivante :
    
    #Si nbCaseLigne est un multiple de 3
    if (nbCaseLigne % 3 == 0) :
        
        #Voir la table de Valeur
        z = nbCaseLigne / 3
    
    #S'il est 1 entier plus grand qu'un multiple de 3.
    elif (nbCaseLigne % 3 == 1) :
        
        #On soustrait nbCaseLigne de son modulo (1)
        z = ((nbCaseLigne - (nbCaseLigne % 3)) / 3)   #Ceci trouve la valeur entière correspondant à z
        
        z +=  0.25 #On ajoute la partie décimale
        
    #S'il est 2 entiers plus grand qu'un multiple de 3.
    elif (nbCaseLigne % 3 == 2) :
        
        #On soustrait nbCaseLigne de son modulo (1)
        z = ((nbCaseLigne - (nbCaseLigne % 3)) / 3) #Ceci trouve la valeur entière correspondant à z
        
        z +=  0.6 #On ajoute la partie décimale
    
    return z #On retourne la valeur de z



#Entrée : Le tableau de jeu (liste 2D)
#Rôle   : Vérifier s'il y a un gagnant en diagonale.
#Sortie : un booléean
def diagonale(case) :
    
    
    #Cas 1 : 
    if case[0][0] == 'X' or case[0][0] == 'O' : 
        
        #On répète aussi longtemps que 
        i = 0
        while i < len(case[0]) :
            
            if case[i][i] == case[0][0] :
                
                #Pour qu'il puisse vérifier toutes les cases avant de donner un gagnant.
                if i == (len(case[0]) - 1) :
                    
                    return True #Il y a un gagnant en diagonale
                
                i += 1; #On incrémente i de 1
                 
              #Si je mets un else à la place, il n'entre pas dans le else
            elif case[i][i] != case[0][0] :
                 #(On ne retourne pas de Faux, car il faut vérifier l'autre diagonale.)
                #break #Ceci va permettre de quitter la boucle
                i = len(case[0]) #Pour éviter d'utiliser break
            
            
     #Cas 2 (La diagonale inversée): 
    if case[0][-1] == 'X' or case[0][-1] == 'O' : 
        
        #On répète aussi longtemps que 
        i = 0
        j = -1 
        
        #Si aucune diagonale, retourne None au lieu d'aller essayer la dernière condition (else)
        while i < len(case[0]) : 
            
            if case[i][j] == case[0][-1] :
                
                #Pour qu'il puisse vérifier toutes les cases avant de donner un gagnant.
                if i == (len(case[0]) - 1) :
                    
                    return True #Il y a un gagnant en diagonale
                
                i += 1; #On incrémente i de 1
                j -= 1; #On diminue j de 1
           
            
            #Si je mets un else à la place, il n'entre pas dans le else
            elif case[i][j] != case[0][-1]:
                
                #break #Ceci va permettre de quitter la boucle
                #i = len(case[0])
                return False #(le faire ici pourrait éviter le return None du while)
                
    
    #Si différent    
    else : 

        return False #Les cases des diagonales ne sont pas tous les mêmes.
    

#Entrée : un tableau représentant les cases du jeu
#Rôle : Vérifier si quelqu'un gagne horizontalement
#Sortie : un booléan (Vrai s'il y a un gagnant, faux sinon)
def gagnant_Horizontale(interface):
    for ligne in interface:
        rep= set(ligne)
        if (len(rep) == 1) and ligne[0] !='_':
            #print("Gagnant(e)")
            #print(ligne)
            return True
         
    else : 

        return False    

    
def Gagnant_Verticale(interface) : 

    for col in range(len(interface)):
        liste_elements = []

        for ligne in interface:
            liste_elements.append(ligne[col])

        if liste_elements.count(liste_elements[0]) == len(liste_elements) and liste_elements[0] != '_':
                return True
    else : 
        return False
    
    
def Verif_Gagnant(case) :
    
    if diagonale(case) :
        
        return True
    
    elif gagnant_Horizontale(case) :
        
        return True
    
    elif Gagnant_Verticale(case) :
        
        return True
    
    else :
        
        return False
        
        
#Entrée : Aucune
#Rôle : Initialiser (dessiner) les cases de jeu
#Sortie : Les cases
def tableDeJeu(nbCaseLigne) :
    #J'ai modifié la  fonction pour qu'elle puisse adapter la génération des cases selon le nombre de cases
    
    #On crée une liste (initialiser avec des valeurs = 0 pour pouvoir donner une taille) nbCaseLigne est exposé en 2, car ça donne le nombre total de cases
    rect = [0] * (nbCaseLigne**2)

    i2 = 0 #Un compteur
    j = 0 # Variable pour variation de positionnement X
    k = 0 # Variable pour variation de positionnement Y
    
    #z sert également à réduire la distance entre les cases
    z = CalculProportionCase(nbCaseLigne) #Variable pour variation de tailles des grilles
    #note la valeur de z maximal est 450. À partir de 451, les cases sont trop petite et il affiche un immense carré blanc... ;) 
    #(mais 450 cases font une grille de jeu extrêmement petite.)
    
    
    
    for i in rect :
            
        #Si le nombre de colonne est dépassé, on remet à 0 pour commencer une nouvelle colonne et on change la rangée
        if j >= nbCaseLigne : 
            j = 0 #On réinitialise j
            k += 1 #On incrémente k pour descendre d'une rangée dans le tableau
                                
        #On dessine la case dans le tableau rect        
        rect[i2] = pygame.draw.rect(fenetre, blanc, (45 + (175 / z) * j,25 + (175 / z) * k, (150 / z ),(150 / z)))
        j += 1  #On augmente le facteur de position de j pour changer de cases
        i2 += 1 #On incrémente pour changer la position de liste à laquelle on veut que la nouvelle case soit dessiné.
    
    return rect, z

#Entrée : Aucune
#Rôle : Réinitialiser les variables essentielles au jeu
#Sortie : Les variables Réinitialisées
def Reinitialiser_Variable(nbCaseLigne) :
    termine = False
    tourOrdinateur = False
    case_disponible = [True] * (nbCaseLigne**2) #Va initialiser une liste qui va contenir le disponibilité de chaque case.
    compteurCasePleine = 0
    
    #Ceci va contenir les différentes case que l'ordinateur peut jouer
    choixOrdi = choixCaseOrdi(nbCaseLigne)
    
    #ChoixOrdi2 va être utile pour effacer (plus tard) des valeurs dans choixOrdi
    choixOrdi2 = choixOrdi.copy()
    
    return termine, tourOrdinateur, case_disponible, compteurCasePleine, choixOrdi, choixOrdi2


#Entrée : Une liste contenant 9 emplacement et la variable de proportionnalité z
#Rôle : Redessiner la grille de jeu
#Sortie : Aucune
def Redessine(case, z, nbCaseLigne) :
    
    
    j = 0
    k = 0 #Initialisation des compteurs pour les boucles suivantes

    #Va boucler tant que K sera plus petit que le nombre de ligne
    while k < nbCaseLigne :

        #Va boucler tant que j sera plus ptit que le nombre de colonne (dans Tic-Tac-Toe, nombre de lignes = nombre de colonnes)
        while j < nbCaseLigne :



            
            #Redessiner le X
            if case[j][k] == 'X' :

                #Note : Malheureusement, la méthode draw.line n'accepte pas les float. En convertissant en int, cela réduit la précision du positionnement et fait que les X sont de moins en moins centré sur les cases s'il y en a beaucoup
                pygame.draw.line(fenetre, rouge, [math.ceil(75/z + 175/z * j), math.ceil(50/z + 175/z * k)], [math.ceil(165/z + 175/z * j), math.ceil(150/z + 175/z * k)], math.ceil(25/z))
                #C'est le même principe qui est efféctué au rangée et au colonne.
                pygame.draw.line(fenetre, rouge, [math.ceil(165/z + 175/z * j), math.ceil(50/z + 175/z * k)], [math.ceil(75/z + 175/z * j), math.ceil(150/z + 175/z * k)], math.ceil(25/z))   
                #Le X est dessiné à l'aide de dux lignes
               

           #Redessiner les O
            elif case[j][k] == 'O':
                 # (120,120) le centre du circle, 50 rayon ) 
                #Note : Malheureusement, la méthode draw.circle n'accepte pas les float. En convertissant en int, cela réduit la précision du positionnement et fait que les O sont de moins en moins centré sur les cases s'il y en a beaucoup
                #On dessine deux cercles, le plus petit aura la couleur de la case pour donner l'illusion d'un trou pour  former la lettre O
                pygame.draw.circle(fenetre, vert, (math.ceil(120/z + 175/z * j), math.ceil(100/z + 175/z * k)), math.ceil(50/z))
                pygame.draw.circle(fenetre, blanc, (math.ceil(120/z + 175/z * j), math.ceil(100/z + 175/z * k)), math.ceil(40/z))
           

            j += 1 #On incrémente j de 1
  
        k += 1 #On incrémente k de 1.
        j = 0 #On réinitialise j

        

        
#60 est la taille de la police d'écriture
font = pygame.font.SysFont(None, 60)



#Entrée : Le message, la couleur et les positions en X et en Y
#Rôle : Écrire un message à l'écran
#Sortie : Aucune
def message_to_screen(msg, color, position_X, position_Y) :
    screen_text = font.render(msg, True, color)
    fenetre.blit(screen_text, [position_X, position_Y])

#Entrée : La fonction pour afficher un message à l'écran
#Rôle : Créer un interface Menu de jeu
#Sortie : Un bool (le choix du joueur (X ou O))
def Menu(message_to_screen) :
    termine = False
    #Dessiner deux rectangles Blanc
    rec_1 = pygame.draw.rect(fenetre, blanc, (75,300,150,150))
    rec_2 = pygame.draw.rect(fenetre, blanc, (375,300,150,150))
    
    #Dessiner un rectangle qui va contenir l'affichage de l'entrée utilisateur.
    rec_input = pygame.draw.rect(fenetre, blanc, (80,120, 440,50))
    
    #Dessiner un X dans le premier Rectangle
    pygame.draw.line(fenetre, rouge, [105,325], [195, 425], 25)
    pygame.draw.line(fenetre, rouge, [195,325], [105, 425], 25)
    
    #Dessiner Un O dans le deuxième rectangle
    pygame.draw.circle(fenetre, vert, (450, 375), 50)
    pygame.draw.circle(fenetre, blanc, (450, 375), 40)
    
    #Afficher ce message à l'écran
    message_to_screen("Veuillez choisir X ou O!", blanc, 70, 180)
      
    #On demande seulement la taille d'une ligne, car la taille de la colonne est OBLIGATOIREMENT de la même longueur.
    message_to_screen("Veuillez entrer la taille ", blanc, 50, 30)
    message_to_screen("d'une ligne du jeu (>= 3):", blanc, 50, 70)
    
    
    #Permet de pouvoir entrer du texte
    pygame.key.set_text_input_rect(rec_input)
    text = ''
                        
    
    pygame.display.flip()
    horloge = pygame.time.Clock()
    
    while termine == False:
        #event comme lic de souris 
        #pygame.event.get() est un list
        for event in pygame.event.get():
            
            if event.type == pygame.QUIT:
                #true monte whie met termine = True, et attend pour 20 second 
                termine = True
            
            #Si le joueur clic sur sa souris
            if event.type == pygame.MOUSEBUTTONUP:
                pos = pygame.mouse.get_pos()
                #print(pos)
                
                #S'il appuie sur la 1ère case
                if rec_1.collidepoint(pos) :
                
                    try :
                        #On essaie de convertir le texte en chiffre
                        nbCaseLigne = int(text)
                    
                    except : 
                        
                        #S'il y a un erreur, on demande d'entre en chiffre + grand que 3
                        message_to_screen("Veuillez entrer un chiffre", rouge, 70, 220)
                        message_to_screen("plus grand que 3!", rouge, 70, 260)
                        pygame.display.flip()
                
                    else : 
                        
                        #Si le chiffre est trop petit pour créer une grille convenable
                        if nbCaseLigne < 3: 
                            
                            #On demande seulement la taille d'une ligne, car la taille de la colonne est OBLIGATOIREMENT de la même longueur.
                            message_to_screen("Veuillez entrer un chiffre", rouge, 70, 220)
                            message_to_screen("plus grand que 3!", rouge, 70, 260)
                            pygame.display.flip()
                            
                        else :
                            termine = True
                            fenetre.fill(noir)
                            return True, nbCaseLigne #True pour les X
            
            
                #S'il appuie sur la deuxième case
                elif rec_2.collidepoint(pos) :
                
                
                    try :
                        #On essaie de convertir le texte en chiffre
                        nbCaseLigne = int(text)
                    
                    except : 
                        
                        #S'il y a un erreur, on demande d'entre en chiffre + grand que 3
                        message_to_screen("Veuillez entrer un chiffre", rouge, 70, 220)
                        message_to_screen("plus grand que 3!", rouge, 70, 260)
                        pygame.display.flip()
                
                    else : 
                        
                        #Si le chiffre est trop petit pour créer une grille convenable
                        if nbCaseLigne < 3: 

                            message_to_screen("Veuillez entrer un chiffre", rouge, 70, 220)
                            message_to_screen("plus grand que 3!", rouge, 70, 260)
                            pygame.display.flip()
                            
                        else :
                            
                            termine = True
                            fenetre.fill(noir)
                            return False, nbCaseLigne #False pour les O

            
            if event.type == pygame.KEYDOWN:

                
                if event.key == pygame.K_BACKSPACE:
                    text = text[:-1]
                    pygame.draw.rect(fenetre, blanc, (80,120, 440,50)) #On repeint l'emplacement
                    pygame.display.flip()
        
                    
                else:
                    text += event.unicode
                    
                message_to_screen(text, noir, 80, 125) #Affiche l'entrée utilisateur
                #dessiner
                pygame.display.flip()

                    
                    
                    
#Entrée : Fonction pour afficher un message et les 2 messages à afficher 
#Rôle : Interface pour savoir si l'utilisateur veut Quitter ou continuer ( ou recommencer) à jouer
#Sortie : Un bool (True pour continuer, False pour quitter)
def Fin_De_Jeu(message_to_screen, msg, msg2, msg3) :
    termine = False
    #Dessiner deux rectangles Blanc
    rec_1 = pygame.draw.rect(fenetre, blanc, (75,300,150,150))
    rec_2 = pygame.draw.rect(fenetre, blanc, (375,300,150,150))
    
    #Affichage de messages
    message_to_screen(msg, blanc, 150, 150)
    message_to_screen(msg2, blanc, 100, 200)
    message_to_screen(msg3, blanc, 100, 50)
    message_to_screen("Oui", vert, 110, 355)
    message_to_screen("Non", rouge, 410, 355)
    
    pygame.display.flip()
    horloge = pygame.time.Clock()

    while termine == False:
        #event comme lic de souris 
        #pygame.event.get() est un list
        for event in pygame.event.get():
            
            if event.type == pygame.QUIT:
                #true monte whie met termine = True, et attend pour 20 second 
                termine = True
            
            if event.type == pygame.MOUSEBUTTONUP:
                pos = pygame.mouse.get_pos()
                #print(pos)
                
                if rec_1.collidepoint(pos) :
                
                    
                    termine = True
                    fenetre.fill(noir)
                    return True #True pour recommencer
            
                elif rec_2.collidepoint(pos) :
                
                    termine = True
                    fenetre.fill(noir)
                    return False #False pour quitter

                
                
                
#Titre
pygame.display.set_caption("Tic-Tac-Toe!")


#Entrée : Un bool, 2 fonctions et 2 lists
#Rôle : Permettre de jouer au jeu
#Sortie : Un bool
def Game(rect_circle_switch, IA_Play, switcher, nbCaseLigne):
    rect, z = tableDeJeu(nbCaseLigne) #On crée le tableau de jeu et on prend la valeur de proportion pour modifier la grosseur des X et des O
    
    case = [['_'] * nbCaseLigne for i in range(nbCaseLigne)] #Ceci va générer un tableau qui contient les cases. Au départ les cases sont vide.
    
    # reinitialier les variable
    termine, tourOrdinateur, case_disponible, compteurCasePleine, choixOrdi, choixOrdi2 = Reinitialiser_Variable(nbCaseLigne)

    
    #à quelle vitesse l'écran met à jour
    # vitesse de jeu,
    #attention a C de Clock
    horloge = pygame.time.Clock()

    #boucle jusqu'à ce que l'utilisateur décide de fermer la fenêtre
    while termine == False:
        #event comme lic de souris 
        #pygame.event.get() est un list
        for event in pygame.event.get():
            if event.type == pygame.QUIT:
                #true monte whie met termine = True, et attend pour 20 second 
                termine = True

            #redémarrage du tableau de jeu
            if event.type == pygame.KEYDOWN:
                
                #J'ai garder cette condition pour pouvoir donner un sens au fonction créée à la question F.
                #Et peut-être qu'elle vous sera utile dans vos tests
                #Si le joueur appuie sur la touche Enter
                if event.key == pygame.K_RETURN:

                    #Réinitialiser le tableau de jeu
                    rect, z = tableDeJeu(nbCaseLigne)

                    # reinitialier les variable
                    termine, tourOrdinateur, case_disponible, compteurCasePleine, choixOrdi, choixOrdi2 = Reinitialiser_Variable(nbCaseLigne)
                
                #S'il appuie sur la touche q
                elif event.key == pygame.K_q :
                    fenetre.fill(noir)  #On repeint tout en noir
                    #Pour demander si l'utilisateur veut quitter. Nonee est retournée à l'emplacement de msg, car on n'envoie pas de message pour cette position
                    termine = Fin_De_Jeu(message_to_screen, None, "Voulez-vous quitter?", None)
                   
                    #S'il veut continuer à jouer
                    if termine == False :
                        fenetre.fill(noir) #On repeint tout en noir
                        rect, z = tableDeJeu(nbCaseLigne) # On replace les cases
                        Redessine(case, z, nbCaseLigne) #On redessine les X et les O
                        
                    #Si termine =  True, le programme va sortir des boucles pour pouvoir quitter

                    
                    
        #LOGIQUE CEST ICI
            #Si le joueur clic avec la souris ou que c'est le tour de l'ordinateur
            if (event.type == pygame.MOUSEBUTTONUP) or  (tourOrdinateur == True) :
                
                #Si c'est le tour du joueur 
                if (tourOrdinateur == False) : 
                    #La variable pos va être la affecté de la position de la souris
                    pos = pygame.mouse.get_pos()
                    #print(pos)
                    
                #Si c'est le tour de L'ordinateur    
                elif (tourOrdinateur == True) :
                    #La varriable pos va être affecté du chois hasardeux de l'ordinateur
                    pos = IA_Play(choixOrdi)
                    #print(f"position ordi : {pos}")
                    
                  
                i = 0
                j = 0
                k = 0 #Initialisation des compteurs pour les boucles suivantes

                #Va boucler tant que K sera plus petit que le nombre de ligne
                while k < nbCaseLigne :
                    
                    #Va boucler tant que j sera plus ptit que le nombre de colonne (dans Tic-Tac-Toe, nombre de lignes = nombre de colonnes)
                    while j < nbCaseLigne :

                        #Si la position est dans la case et que la case est libre
                        
                        if rect[i].collidepoint(pos) and case_disponible[i]:
                            #Si le joueur ou l'Ordinateur à les X 
                            if rect_circle_switch == True :
                                
                                #Note : Malheureusement, la méthode draw.line n'accepte pas les float. En convertissant en int, cela réduit la précision du positionnement et fait que les X sont de moins en moins centré sur les cases s'il y en a beaucoup
                                pygame.draw.line(fenetre, rouge, [math.ceil(75/z + 175/z * j), math.ceil(50/z + 175/z * k)], [math.ceil(165/z + 175/z * j), math.ceil(150/z + 175/z * k)], math.ceil(25/z))
                                #C'est le même principe qui est efféctué au rangée et au colonne.
                                pygame.draw.line(fenetre, rouge, [math.ceil(165/z + 175/z * j), math.ceil(50/z + 175/z * k)], [math.ceil(75/z + 175/z * j), math.ceil(150/z + 175/z * k)], math.ceil(25/z))   
                                #Le X est dessiné à l'aide de dux lignes

                                #Ceci va permettre l'alternance entre le joueur et l'ordinateur
                                tourOrdinateur = switcher(tourOrdinateur)
                                rect_circle_switch = switcher(rect_circle_switch) #Pour que l'ordinateur ne joue pas le même symbole que le joueur
                                #rect_circle_switch = False
                                case[j][k] = 'X'  #On dit à la liste que cette case est un X

                            #Si le joueur ou l'Ordinateur à les O 
                            else:
                                 # (120,120) le centre du circle, 50 rayon ) 
                                #Note : Malheureusement, la méthode draw.circle n'accepte pas les float. En convertissant en int, cela réduit la précision du positionnement et fait que les O sont de moins en moins centré sur les cases s'il y en a beaucoup
                                #On dessine deux cercles, le plus petit aura la couleur de la case pour donner l'illusion d'un trou pour  former la lettre O
                                pygame.draw.circle(fenetre, vert, (math.ceil(120/z + 175/z * j), math.ceil(100/z + 175/z * k)), math.ceil(50/z))
                                pygame.draw.circle(fenetre, blanc, (math.ceil(120/z + 175/z * j), math.ceil(100/z + 175/z * k)), math.ceil(40/z))

                                #Ceci va permettre l'alternance entre le joueur et l'ordinateur
                                tourOrdinateur = switcher(tourOrdinateur)
                                rect_circle_switch = switcher(rect_circle_switch)
                                #rect_circle_switch = True
                                case[j][k] = 'O' #On dit à la liste que cette case est un O
                            case_disponible[i] = False  #Puisque la case n'est plus disponnible
                            compteurCasePleine += 1 #Pour Augmenter la valeur du compteur (python n'accepte pas la notation ++)
                            
                            choixOrdi.remove(choixOrdi2[i]) #On efface la case des possibilités de jeu de l'ordinateur. Cela devrait légèrement augmenter sa vitesse pour jouer.
                        
                        
                        j += 1 #On incrémente j de 1
                        i += 1 #On incrémente i de 1
                    k += 1 #On incrémente k de 1.
                    j = 0 #On réinitialise j
                
           
            #Va s'exécuter s'il y a une possibilité qu'il y ait un gagnant et va vérifier s'il y a un gagnant
            if (compteurCasePleine >= nbCaseLigne) and Verif_Gagnant(case) :
                
                termine = True #Pour sortir de cette boucle
                pygame.display.flip()#dessiner le dernier X ou O
                time.sleep(3) #On attend 3 secondes pour permettre à l'utilisateur de voir le tableau de jeu
                fenetre.fill(noir) #On repeint l'écran en noir
                
                #Si l'ordinateur perd (La variable tour ordinateur a été changé plus haut dans le code ce qui fait qu'il faut regarder l'inverse)
                if tourOrdinateur == True :
                    manyGame = Fin_De_Jeu(message_to_screen, "Fin de la partie!", "Voulez-vous rejouer?", "Vous avez gagné!") #Afficher un message
                    return manyGame #Un bool (true or false qui va permettre de savoir s'il veut rejouer)
                
                else :
                    
                    manyGame = Fin_De_Jeu(message_to_screen, "Fin de la partie!", "Voulez-vous rejouer?", "L'ordinateur a gagné!") #Afficher un message
                    return manyGame #Un bool (true or false qui va permettre de savoir s'il veut rejouer)
        
            #Si toute les cases du jeu sont prises (Fin de la partie)
            elif compteurCasePleine == nbCaseLigne**2 :
                termine = True #Pour sortir de cette boucle
                pygame.display.flip()#dessiner le dernier X ou O
                time.sleep(3) #On attend 3 secondes pour permettre à l'utilisateur de voir le tableau de jeu
                fenetre.fill(noir) #On repeint l'écran en noir
                manyGame = Fin_De_Jeu(message_to_screen, "Fin de la partie!", "Voulez-vous rejouer?", "Match Nulle!") #Afficher un message
                return manyGame #Un bool (true or false qui va permettre de savoir s'il veut rejouer)
            
        #dessiner
        pygame.display.flip()

        # si on dessin le rectangle ici, on voit rien

        #20 images par seconde, 10 moin vite que 20        
        horloge.tick(20)
        

manyGame = True
#On boucle le jeu jusqu'à ce que l'utilisateur ne veuille plus jouer
while manyGame == True : 
    rect_circle_switch, nbCaseLigne = Menu(message_to_screen) #On affiche le menu et on récolte son choix
    manyGame = Game(rect_circle_switch, IA_Play, switcher, nbCaseLigne) #On lance la partie
#     manyGame = Fin_De_Jeu(message_to_screen, "Fin de la partie!", "Voulez-vous rejouer?")
pygame.quit()
```

```python
#Algorithme original du TP2

#Entrée : Aucune
#Rôle : Initialiser (dessiner) les cases de jeu
#Sortie : Les 9 cases
def tableDeJeu(nbCaseLigne) :
    
    #J'ai ajouté 20 à vos valeurs pour que les carrées aient l'air plus centrée
    # on ajout 175 à x donc on commence par 45, 45+175 = 220, 220+175=  395
    rec_1 = pygame.draw.rect(fenetre, blanc, (45,25,150,150))
    # (255,255,255) couleur blanc, on change x a 200 
    rec_2 = pygame.draw.rect(fenetre, (255,255,255), (220,25,150,150))
    rec_3 = pygame.draw.rect(fenetre, blanc, (395,25,150,150))
    #y starts at 200
    rec_4 = pygame.draw.rect(fenetre, blanc, (45,200,150,150))
    rec_5 = pygame.draw.rect(fenetre, blanc, (220,200,150,150))
    rec_6 = pygame.draw.rect(fenetre, blanc, (395,200,150,150))
    #y starts at 375
    rec_7 = pygame.draw.rect(fenetre, blanc, (45,375,150,150))
    rec_8 = pygame.draw.rect(fenetre, blanc, (220,375,150,150))
    rec_9 = pygame.draw.rect(fenetre, blanc, (395,375,150,150))
    
    return rec_1, rec_2, rec_3, rec_4, rec_5, rec_6, rec_7, rec_8, rec_9





#Le nouvel algorithme (Il est beaucoup plus gros, mais il permet de faire plusieurs styles de parties)



#Entrée : Le nombre de case par ligne   
#Rôle : Fonction qui calcul le z de la fonction tableDeJeu()    
#Sortie : le chiffre calculé par la fonction    
def CalculProportionCase(nbCaseLigne) :
    
    #Table des valeurs possible (Je les ai calculé en faisant des tests)
    # x | 3  |  4   |  5  | 6 |  7   |  8  | 9 |  10  |...
    # y | 1 | 1.25 | 1.6 | 2 | 2.25 | 2.6 | 3 | 3.25 | ...
    
    #À l'aide de cette table de valeur, je peux créer la fonction suivante :
    
    #Si nbCaseLigne est un multiple de 3
    if (nbCaseLigne % 3 == 0) :
        
        #Voir la table de Valeur
        z = nbCaseLigne / 3
    
    #S'il est 1 entier plus grand qu'un multiple de 3.
    elif (nbCaseLigne % 3 == 1) :
        
        #On soustrait nbCaseLigne de son modulo (1)
        z = ((nbCaseLigne - (nbCaseLigne % 3)) / 3)   #Ceci trouve la valeur entière correspondant à z
        
        z +=  0.25 #On ajoute la partie décimale
        
    #S'il est 2 entiers plus grand qu'un multiple de 3.
    elif (nbCaseLigne % 3 == 2) :
        
        #On soustrait nbCaseLigne de son modulo (1)
        z = ((nbCaseLigne - (nbCaseLigne % 3)) / 3) #Ceci trouve la valeur entière correspondant à z
        
        z +=  0.6 #On ajoute la partie décimale
    
    return z #On retourne la valeur de z
        
        
        
#Entrée : Aucune
#Rôle : Initialiser (dessiner) les cases de jeu
#Sortie : Les cases
def tableDeJeu(nbCaseLigne) :
    #J'ai modifié la  fonction pour qu'elle puisse adapter la génération des cases selon le nombre de cases
    
    #On crée une liste (initialiser avec des valeurs = 0 pour pouvoir donner une taille) nbCaseLigne est exposé en 2, car ça donne le nombre total de cases
    rect = [0] * (nbCaseLigne**2)

    i2 = 0 #Un compteur
    j = 0 # Variable pour variation de positionnement X
    k = 0 # Variable pour variation de positionnement Y
    
    #z sert également à réduire la distance entre les cases
    z = CalculProportionCase(nbCaseLigne) #Variable pour variation de tailles des grilles
    #note la valeur de z maximal est 450. À partir de 451, les cases sont trop petite et il affiche un immense carré blanc... ;) 
    #(mais 450 cases font une grille de jeu extrêmement petite.)
    
    
    
    for i in rect :
            
        #Si le nombre de colonne est dépassé, on remet à 0 pour commencer une nouvelle colonne et on change la rangée
        if j >= nbCaseLigne : 
            j = 0 #On réinitialise j
            k += 1 #On incrémente k pour descendre d'une rangée dans le tableau
                                
        #On dessine la case dans le tableau rect        
        rect[i2] = pygame.draw.rect(fenetre, blanc, (45 + (175 / z) * j,25 + (175 / z) * k, (150 / z ),(150 / z)))
        j += 1  #On augmente le facteur de position de j pour changer de cases
        i2 += 1 #On incrémente pour changer la position de liste à laquelle on veut que la nouvelle case soit dessiné.
    
    return rect
```

## La fonction pour vérifier un gagnant

### (utile ici pour tester les possibilités de jeu en 3X3)

```python
#Entrée : Le tableau de jeu (liste 2D)
#Rôle   : Vérifier s'il y a un gagnant en diagonale.
#Sortie : un booléean
def diagonale(case) :
    
    
    #Cas 1 : 
    if case[0][0] == 'X' or case[0][0] == 'O' : 
        
        #On répète aussi longtemps que 
        i = 0
        while i < len(case[0]) :
            
            if case[i][i] == case[0][0] :
                
                #Pour qu'il puisse vérifier toutes les cases avant de donner un gagnant.
                if i == (len(case[0]) - 1) :
                    
                    return True #Il y a un gagnant en diagonale
                
                i += 1; #On incrémente i de 1
                 
              #Si je mets un else à la place, il n'entre pas dans le else
            elif case[i][i] != case[0][0] :
                 #(On ne retourne pas de Faux, car il faut vérifier l'autre diagonale.)
                #break #Ceci va permettre de quitter la boucle
                i = len(case[0]) #Pour éviter d'utiliser break
            
            
     #Cas 2 (La diagonale inversée): 
    if case[0][-1] == 'X' or case[0][-1] == 'O' : 
        
        #On répète aussi longtemps que 
        i = 0
        j = -1 
        
        #Si aucune diagonale, retourne None au lieu d'aller essayer la dernière condition (else)
        while i < len(case[0]) : 
            
            if case[i][j] == case[0][-1] :
                
                #Pour qu'il puisse vérifier toutes les cases avant de donner un gagnant.
                if i == (len(case[0]) - 1) :
                    
                    return True #Il y a un gagnant en diagonale
                
                i += 1; #On incrémente i de 1
                j -= 1; #On diminue j de 1
           
            
            #Si je mets un else à la place, il n'entre pas dans le else
            elif case[i][j] != case[0][-1]:
                
                #break #Ceci va permettre de quitter la boucle
                #i = len(case[0])
                return False #(le faire ici pourrait éviter le return None du while)
                
    
    #Si différent    
    else : 

        return False #Les cases des diagonales ne sont pas tous les mêmes.
    

#Entrée : un tableau représentant les cases du jeu
#Rôle : Vérifier si quelqu'un gagne horizontalement
#Sortie : un booléan (Vrai s'il y a un gagnant, faux sinon)
def gagnant_Horizontale(interface):
    for ligne in interface:
        rep= set(ligne)
        if (len(rep) == 1) and ligne[0] !='_':
            #print("Gagnant(e)")
            #print(ligne)
            return True
         
    else : 

        return False    

    
def Gagnant_Verticale(interface) : 

    for col in range(len(interface)):
        liste_elements = []

        for ligne in interface:
            liste_elements.append(ligne[col])

        if liste_elements.count(liste_elements[0]) == len(liste_elements) and liste_elements[0] != '_':
                return True
    else : 
        return False
    
    
def Verif_Gagnant(case) :
    
    if diagonale(case) :
        
        return True
    
    elif gagnant_Horizontale(case) :
        
        return True
    
    elif Gagnant_Verticale(case) :
        
        return True
    
    else :
        
        return False
    
    
case = [ ['O','X','O'],
         ['_','X','_'],
         ['X','O','X'],] 

print(Verif_Gagnant(case))
```
